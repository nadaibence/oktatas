```python
import pandas as pd
import numpy as np
import sqlalchemy
```

```python
df = pd.DataFrame([[1,2,3]], columns=['A','B','C'])
```

```python
connection_string = 'postgresql://postgres:postgres@postgres'
conn_engine = sqlalchemy.create_engine(connection_string)
```

```python
df.to_sql('test', index=False, if_exists='replace', con=conn_engine)
```

```python
con = conn_engine.connect()
```

```python
asd = con.execute('select * from test')
```

```python
for i in asd:
    print(i)
```
